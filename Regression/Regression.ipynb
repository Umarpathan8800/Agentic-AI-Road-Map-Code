### Linear Regression with 1 Variable --> ind = Exp & dep = Salary


```python
df_1feat = data[['Exp', 'Salary']]
df_1feat
```

```python
df_1feat.plot(x= 'Exp', y= 'Salary', style= 'o')
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.savefig('Linear_Regession_1.png')
plt.show()
```

```python
### Segregate the Dependent and Independent Variables:

x= pd.DataFrame(df_1feat['Exp'])
y= pd.DataFrame(df_1feat['Salary'])
```

```python
## Divide thee dataset into train and test:

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state= 42)
```

```python
print(f"X Train shape is: {x_train.shape}")
print(f"Y Train shape is: {y_train.shape}")
print("-"*30)
print(f"X Test shape is: {x_test.shape}")
print(f"Y Test shape is: {y_test.shape}")


```

```python
## Create a Model:

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
## Fit the model
```

```python
model.fit(x_train, y_train)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
### Predications:

y_pred = model.predict(x_test)
y_pred = pd.DataFrame(y_pred, columns= ['Salary'])
y_pred.size
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
## Calculating Errors:

print(f"MAE: {metrics.mean_absolute_error(y_test, y_pred)}")
print(f"MSE: {metrics.mean_squared_error(y_test, y_pred)}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}")
```

```python
print(f"RSquared : {metrics.r2_score(y_test, y_pred)}")
```

Multiple Linear Regression

```python
### Segregate the Dependent (all) and Independent Variables:

x = pd.DataFrame(data.iloc[:,:-1])
y = pd.DataFrame(data.iloc[:,-1])

```

```python
data.head(2)
```

```python
## Divide thee dataset into train and test:

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state= 42)

print(f"X Train shape is: {x_train.shape}")
print(f"Y Train shape is: {y_train.shape}")
print("-"*30)
print(f"X Test shape is: {x_test.shape}")
print(f"Y Test shape is: {y_test.shape}")

```

```python
## Create a Model:

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
## Fit the model

x_train_encoded = pd.get_dummies(x_train, columns=['Dept', 'Rating', 'City'], drop_first=True)
x_test_encoded = pd.get_dummies(x_test, columns=['Dept', 'Rating', 'City'], drop_first=True)

model.fit(x_train_encoded, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
### Predications:

y_pred = model.predict(x_test_encoded)
y_pred = pd.DataFrame(y_pred, columns= ['Salary'])
y_pred.size
```

```python
y_pred
```

```python
## Calculating Errors:

print(f"MAE: {metrics.mean_absolute_error(y_test, y_pred)}")
print(f"MSE: {metrics.mean_squared_error(y_test, y_pred)}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}")
```
