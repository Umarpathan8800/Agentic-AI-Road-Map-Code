### List vs Numpy array

```python
py_list = [1,2,3]
print("Python list multi..", py_list * 3)
```

```python
np_array = np.array([1,2,3]) # elements wise mutliplication...
print("Python list multi..", np_array * 3)
```

```python
import time
start = time.time()
py_list = [i * 2 for i in range(1000000)]
print("\n List operation time: ", time.time()-start)


start = time.time()
np_array = np.arange(1000000) 
print("\n Numpy operation time: ", time.time()-start)
```

### Vector, Matrix and Tensor


```python
vector = np.array([1,2,3])
print("vector:", vector)


matrix = np.array([[1,2,3],
                   [4,5,6,]])
print("Matrix:", matrix)

tensor = np.array([[
                   [1,2],[3,4]],
                   [[5,6], [7,8]
                          ]])
print("tensor:", tensor)
```

### Array Properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Shape ", arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("DType ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Array:", arr)

reshaped = arr.reshape((4,3))
print("\n Reshaped Array: \n",reshaped)

flattend = reshaped.flatten()
print("\n Flattend Array:", flattend)

# ravel (returns the view, instead of copy)
raveled = reshaped.ravel()
print("\n Raveled Array:", raveled)
```

### Transpose 

```python
transpose = reshaped.T
print("\n Transposed Array: ", transpose)
```
