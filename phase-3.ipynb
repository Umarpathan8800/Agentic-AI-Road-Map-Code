## Advance operation with Busniness examples

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# data structure :[ restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
  [1, 150000, 180000, 220000, 250000],
  [2, 100000, 125000, 155000, 190000],
  [3, 140000, 170000, 210000, 260000],
  [4, 110000, 150000, 185000, 225000],
  [5, 160000, 190000, 230000, 270000]
  ])

print(" === Zomato sales analysis ===")
print("\n Sales data shpe:" , sales_data.shape)
print("\n Sample data for 1st 3 restau:\n ", sales_data[0:3])
```

```python
# total sales per year 
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# Mininum sales per restaurant 
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per restaurant 
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restaurant 
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector Addition: " , vector1 + vector2)
print(" \n Vector Multiplication: " , vector1 * vector2)
print("\n Dot product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
